```python
def find_nearest(arr, target):
    if not arr:
        raise ValueError("Array is empty")

    closest_index = 0
    closest_difference = abs(arr[0] - target)

    for i in range(1, len(arr)):
        difference = abs(arr[i] - target)
        if difference < closest_difference:
            closest_index = i
            closest_difference = difference

    return closest_index
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
def fidelity(rho_1,rho_2):
    norm_1 = np.trace(rho_1)
    norm_2 = np.trace(rho_2)
    F = scipy.linalg.sqrtm(np.dot(np.dot(scipy.linalg.sqrtm(rho_1),rho_2),scipy.linalg.sqrtm(rho_1)))
    return np.trace(F)/np.sqrt(norm_1*norm_2) 
```

```python
def analytical_FT(omega_tau,A,sigma_1,sigma_2,phi_1,phi_2,omega_1,omega_2,E,omega_a,q_ag):
    delta = (omega_1+omega_2-E)*hbar
    epsilon = (E-np.real(omega_a))/np.abs(np.imag(omega_a))    
    sigma = hbar*np.sqrt(sigma_1**2+sigma_2**2)
    sigma_t = np.sqrt(1/sigma_1**2+1/sigma_2**2)/hbar
    t1 = np.exp(-delta**2/(2*sigma**2))
    t2 = np.exp(-sigma_t**2/2*(omega_tau-hbar*omega_2+sigma_2**2/sigma**2*delta)**2)
    t3 = (1/omega_tau + (beta_a-epsilon**(-1))*(q_ag-1j)/(omega_tau-(hbar*E-hbar*omega_a)))
    return A*1j*(t1*t2*t3)

def RK_analytical(X,Y,omega_1,omega_2,omega_a,Gamma,q,sigma_1):
    epsilon_tau = (Y)/Gamma
    epsilon_Ef = (X)/Gamma
    beta= -0.0003
    epsilon_a = X+omega_1+omega_2-omega_a
    xi = omega_2/Gamma*(-beta+1./epsilon_a)
    Delta_probe = (q-1j)*(Y/omega_2*xi-1)
    Delta_ref = (q-1j)* (xi-1)
    term_1 = (epsilon_tau+q+Delta_probe)/(epsilon_tau+1j)
    term_2 = (epsilon_Ef+q+Delta_ref)/(epsilon_Ef+1j)
    modulation = np.exp(-(X)**2/(2*hbar**2*sigma_1**2))*np.exp(-(Y)**2/(2*hbar**2*sigma_1**2))
    return term_1*term_2*modulation
```

```python
def rainbow_kraken_no_plots(params_system,params_protocol,params_simulation, axes_):

    # Unpack parameters
    [params, params_p] = params_system
    
    [v_A1,v_A2,v_sigma_1,  v_sigma_2, v_sigma_2_broad,v_sigma_2_narrow, v_phi_1,v_phi_2,v_omega_1,  v_omega_2, v_omega_2_broad, v_omega_2_narrow, v_E,v_omega_a,v_q_ag] = params 
    [v_A1,v_A2,v_sigma_1_p,v_sigma_2, v_sigma_2_broad,v_sigma_2_narrow, v_phi_1,v_phi_2,v_omega_1_p,v_omega_2, v_omega_2_broad, v_omega_2_narrow, v_E,v_omega_a,v_q_ag] = params_p

    
    VIA_DISCRETE, SPIN_ORBIT, time_probe, time_ref = params_protocol
    [Ef_initial, Ef_final, N_Ef, N_tau, time_limits, N_interpolation] = params_simulation
    
    [ax_time_traces_1, ax_time_traces_2, ax_time_traces_3,
        ax_fourier_transform_1,ax_fourier_transform_2,ax_fourier_transform_3,
        ax_fourier_rescaled_1,ax_fourier_rescaled_2,
        ax_fourier_rescaled_3,ax_fourier_rescaled_4,
        ax_fourier_rescaled_5,ax_fourier_rescaled_6,
        ax_interpolated_1,ax_interpolated_2,ax_interpolated_3,
        ax_interpolated_4,ax_interpolated_5,ax_interpolated_6,
        ax_interpolated_7,ax_interpolated_8,ax_interpolated_9] = axes_     

    #######################################################################################
    # 1. TIME TRACES
    #######################################################################################
    
    
    # Set up grids
    time_range = 2.*time_limits
    grid_Ef = np.linspace(Ef_initial,Ef_final,N_Ef) # Energy grid in angular frequency
    grid_Omegaf = grid_Ef/hbar
    grid_tau = np.linspace(-time_limits,time_limits,N_tau)
    X_Ef,Y_tau = np.meshgrid(grid_Ef,grid_tau)
    X_Omegaf,Y_tau = np.meshgrid(grid_Omegaf,grid_tau)
    
    ################################################################################
    
    # Setup the grids for the measured signal. We contemplate two incoherent signals. 
    Z = signal(X_Omegaf,v_A1,v_A2,v_sigma_1,  v_sigma_2,v_phi_1,v_phi_2,v_omega_1,  v_omega_2,Y_tau,v_omega_a,v_q_ag)
    Z = signal(X_Omegaf,v_A1,v_A2,v_sigma_1,  v_sigma_2,v_phi_1,v_phi_2,v_omega_1,  v_omega_2,Y_tau,v_omega_a,v_q_ag)
    Z_broadband = np.zeros((N_Ef,N_tau),'complex')
    Z_reference = np.zeros((N_Ef,N_tau),'complex')

    Z_p = signal(X_Omegaf,v_A1,v_A2,v_sigma_1_p,v_sigma_2,v_phi_1,v_phi_2,v_omega_1_p,v_omega_2,Y_tau,v_omega_a,v_q_ag)
    Z_p = signal(X_Omegaf,v_A1,v_A2,v_sigma_1_p,v_sigma_2,v_phi_1,v_phi_2,v_omega_1_p,v_omega_2,Y_tau,v_omega_a,v_q_ag)
    Z_broadband_p = np.zeros((N_Ef,N_tau),'complex')
    Z_reference_p = np.zeros((N_Ef,N_tau),'complex')

    #######################################################################################
    # 1.1 Calculate the time traces
    #######################################################################################
    
    # Calculate each of the interactions with each component of the probe
    #for aa in range(N_Ef):
    #    for bb in range(N_tau):
            
            # First statistical mixture
    #        Z_broadband[aa,bb] =  signal(grid_tau[bb]*time_probe,v_A1,v_A2,v_sigma_1,v_sigma_2_broad,v_phi_1,v_phi_2,v_omega_1,v_omega_2_broad,grid_Omegaf[aa],v_omega_a,v_q_ag)
    #        Z_reference[aa,bb] =  signal(grid_tau[bb]*time_ref,v_A1,v_A2,v_sigma_1,v_sigma_2_narrow,v_phi_1,v_phi_2,v_omega_1,v_omega_2_narrow,grid_Omegaf[aa],v_omega_a,v_q_ag)
            
            # Second statistical mixture
    #        Z_broadband_p[aa,bb] = signal(grid_tau[bb]*time_probe,v_A1,v_A2,v_sigma_1_p,v_sigma_2_broad,v_phi_1,v_phi_2, v_omega_1_p,v_omega_2_broad, grid_Omegaf[aa],v_omega_a,v_q_ag)
    #        Z_reference_p[aa,bb] = signal(grid_tau[bb]*time_ref,  v_A1,v_A2,v_sigma_1_p,v_sigma_2_narrow,v_phi_1,v_phi_2,v_omega_1_p,v_omega_2_narrow,grid_Omegaf[aa],v_omega_a,v_q_ag)
    
    # Build 2D grids once (shape: (N_Ef, N_tau))
    Ef_grid, Tau_base = np.meshgrid(grid_Omegaf, grid_tau, indexing="ij")
    Tau_probe = Tau_base * time_probe
    Tau_ref   = Tau_base * time_ref

    # First statistical mixture
    Z_broadband = signal(Tau_probe, v_A1, v_A2, v_sigma_1, v_sigma_2_broad, v_phi_1, v_phi_2,
        v_omega_1, v_omega_2_broad, Ef_grid, v_omega_a, v_q_ag)
    Z_reference = signal(Tau_ref, v_A1, v_A2, v_sigma_1, v_sigma_2_narrow, v_phi_1, v_phi_2,
        v_omega_1, v_omega_2_narrow, Ef_grid, v_omega_a, v_q_ag)

    # Second statistical mixture (spin–orbit variant)
    Z_broadband_p = signal(Tau_probe, v_A1, v_A2, v_sigma_1_p, v_sigma_2_broad, v_phi_1, v_phi_2,
        v_omega_1_p, v_omega_2_broad, Ef_grid, v_omega_a, v_q_ag)
    Z_reference_p = signal(Tau_ref, v_A1, v_A2, v_sigma_1_p, v_sigma_2_narrow, v_phi_1, v_phi_2,
        v_omega_1_p, v_omega_2_narrow, Ef_grid, v_omega_a, v_q_ag)

    
    
    #Z_RK  = (Z_broadband*np.conjugate(Z_reference) + np.conjugate(Z_broadband)*(Z_reference) 
    #         +  Z_broadband*np.conjugate(Z_broadband) + Z_reference*np.conjugate(Z_reference)  )
    #Z_RK_p  = (Z_broadband_p*np.conjugate(Z_reference_p) + np.conjugate(Z_broadband_p)*(Z_reference_p) 
    #         +  Z_broadband_p*np.conjugate(Z_broadband_p) + Z_reference_p*np.conjugate(Z_reference_p)  )
    
    Z_RK  = (Z_broadband*np.conjugate(Z_reference))
    Z_RK_p  = (Z_broadband_p*np.conjugate(Z_reference_p)  )
    
    
    
    
    if (SPIN_ORBIT==0):
        Z_RK_total = Z_RK
    
    if (SPIN_ORBIT==1):
        Z_RK_total = 1/3*Z_RK + 2/3*Z_RK_p
    
    
    if (SYMMETRIZE_TIME_TRACE==1):
        YY = np.transpose(Z_RK_total)
        #N_tau = len(grid_tau)
        YY_positive = YY[int((N_tau-1)/2):N_tau,:]
        YY_adjusted = (np.conjugate(YY_positive))[::-1,:]
        YY_new = YY
        YY_new[0:int((N_tau-1)/2)+1,:] = YY_adjusted
        Z_RK_total = np.transpose(YY_new)
    
    if (NOISE==1):
        mean = 0

        max_intensity_Z = np.amax(np.abs(Z_RK_total.copy()))
        
        # Generate the same shape noise array as Z_RK_total
        noise = np.random.normal(mean, std_dev*max_intensity_Z, Z_RK_total.shape)

        # Add the noise to the original array to create a new array with noise
        Z_RK_total = Z_RK_total + noise
    
    #######################################################################################
    # 1.2 Plot the traces
    #######################################################################################
    
    
    
    #######################################################################################
    # 1.3 Return arrays
    #######################################################################################
    
    array_time_traces  = [grid_Ef, grid_tau, X_Ef, Y_tau, 
                          Z_broadband, Z_reference, Z_broadband_p, Z_reference_p,
                          Z_RK_total]
    plots_time_traces  = [ax_time_traces_1,ax_time_traces_2,ax_time_traces_3]
    return_time_traces = [array_time_traces,plots_time_traces]
    
    #######################################################################################
    # 2. FOURIER TRANSFORM
    #######################################################################################
    
    
    #######################################################################################
    # 2.1 Calculate and plot the Fourier transforms
    #######################################################################################

    # Defines the frequency for the Fourier transform

    s_time_interval=(time_range)/N_tau # define the time interval
    f_range = 1./(2.*s_time_interval)  # frequency range based on the interval
    freq = 2.*math.pi*np.linspace(-f_range,f_range,N_tau)*0.658  # define the frequency array
    X_Ef,Y_freq = np.meshgrid(grid_Ef,freq)
    

    # Define the array to hold the spectra to get the spectra
    Z_RK_FT = Z_RK_total.copy() # Copy the array where the FT will be
    Z_RK_FT[np.isnan(Z_RK_FT)] = 0 # replace nans with zeros

    # Calculate the Fourier transform
    
    for aa in range(len(grid_Ef)):
        Z_RK_FT[aa,:] = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(Z_RK_FT[aa,:]))) 
    Z_RK_FT = np.transpose(Z_RK_FT)
    Z_RK_FT[np.isnan(Z_RK_FT)] = 0
    #Z_RK_FT[freq < 0] = 0
    
    # Define the energy at which we find the resonance, x_a, and the center energy for the XUV+IR pulse

    x_a = (np.real(v_omega_a)+v_omega_2)*hbar
    x_omega = (v_omega_1+v_omega_2)*hbar

    
    #######################################################################################
    # CORRECTION FACTORS
    #######################################################################################
    
    # sigmas are in frequencies
    v_sigma = np.sqrt( v_sigma_1**2 + v_sigma_2_broad**2 )
    v_sigma_t = np.sqrt( 1./v_sigma_1**2 + 1./v_sigma_2_broad**2 )
    
    # Naive correction based on bandwidth of sigma2
    spectrum_correction_intuitive = np.exp(-(Y_freq-v_omega_2_broad*hbar)**2 / (2 * (v_sigma_2_broad*hbar)**2))+0.01
    # Correction based on analytical equation
    spectrum_correction_analytical = 1/(Y_freq+0.001)*( np.exp(-(Y_freq-v_omega_2_broad*hbar 
                      + v_sigma_2_broad**2/v_sigma**2 *(hbar*v_omega_1 + hbar*v_omega_2_broad - X_Ef) )**2 
                                       *v_sigma_t**2/(2*hbar**2) ) 
            * np.exp(-(hbar*v_omega_1 + hbar*v_omega_2_broad - X_Ef)**2/(2*hbar**2*v_sigma**2)) +0.001 )
     
    restore_profile_Fourier =  np.exp(-(  Y_freq + v_omega_1*hbar - X_Ef)**2/(2*hbar**2*v_sigma_1**2) )    
    mod =  np.exp(-(  Y_freq + v_omega_1*hbar - X_Ef)**2/(2*hbar**2*v_sigma_1**2) )    

    
    # Correct with the complicated exponential factor that is derived from the theory
    Z_RK_FT_rs_analytical = Z_RK_FT.copy() / spectrum_correction_analytical
    Z_RK_FT_rs_intuitive  = Z_RK_FT.copy() / spectrum_correction_intuitive

    Z_RK_FT_rs_analytical_mod = Z_RK_FT_rs_analytical.copy() * restore_profile_Fourier 
    Z_RK_FT_rs_intuitive_mod = Z_RK_FT_rs_intuitive.copy() * restore_profile_Fourier 
    
    #######################################################################
    
    if False:
        print('Restricting the wavepacket range')
        # sigmas are in frequencies
        v_sigma = np.sqrt( v_sigma_1**2 + v_sigma_2_broad**2 )
        v_sigma_t = np.sqrt( 1./v_sigma_1**2 + 1./v_sigma_2_broad**2 )

        # Correction based on analytical equation
        spectrum_correction_analytical = 1/(Y_freq+0.001)*( np.exp(-(Y_freq-v_omega_2_broad*hbar 
                          + v_sigma_2_broad**2/v_sigma**2 *(hbar*v_omega_1 + hbar*v_omega_2_broad - X_Ef) )**2 
                                           *v_sigma_t**2/(2*hbar**2) ) 
                * np.exp(-(hbar*v_omega_1 + hbar*v_omega_2_broad - X_Ef)**2/(2*hbar**2*v_sigma**2)) +0.001 )

        v_sigma_F = v_sigma_2_broad
        restore_profile_Fourier_F =  np.exp(-(  Y_freq + v_omega_1*hbar - X_Ef)**2/(2*hbar**2*v_sigma_F**2) )    


        profile_X = ( np.exp( -(X_Ef-hbar*v_omega_1-hbar*v_omega_2_broad)**2/(2*(hbar*v_sigma_1)**2) ) +0.001 )
        restore_X = np.exp( -(X_Ef-hbar*v_omega_1-hbar*v_omega_2_broad)**2/(2*(hbar*v_sigma_F)**2) )

        new_Y = (Z_RK_FT.copy()) / spectrum_correction_analytical*restore_profile_Fourier_F / profile_X * restore_X

        Z_RK_FT_rs_analytical_mod = new_Y.copy()

    ######################################################################
    
    Z_RK_FT_rs_analytical[np.isnan(Z_RK_FT_rs_analytical)] = 0 
    Z_RK_FT_rs_intuitive[np.isnan(Z_RK_FT_rs_intuitive)] = 0 
    Z_RK_FT_rs_analytical_mod[np.isnan(Z_RK_FT_rs_analytical_mod)] = 0 
    Z_RK_FT_rs_intuitive_mod[np.isnan(Z_RK_FT_rs_intuitive_mod)] = 0 
    
    #######################################################################################
    # 2.2 Plot Fourier transform without any correction
    #######################################################################################
    
    
    #######################################################################################
    # 2.3 Return arrays
    #######################################################################################
    
    array_fourier_transform  = [grid_Ef, freq, X_Ef, Y_freq, 
                                Z_RK_FT, Z_RK_FT_rs_analytical, Z_RK_FT_rs_intuitive,
                               spectrum_correction_analytical, spectrum_correction_intuitive]
    plots_fourier_transform  = [ax_fourier_transform_1,ax_fourier_transform_2,ax_fourier_transform_3]
    return_fourier_transform = [array_fourier_transform, plots_fourier_transform]
    
    
    
    #######################################################################################
    # 3. RESCALE FOURIER TRANSFORM
    #######################################################################################
    
    #######################################################################################
    # 3.1 Rescale the axis to get the Fourier transform
    #######################################################################################

    X, Y = X_Ef, Y_freq #np.meshgrid(grid_E*hbar,freq) X_Omegaf

    # This part replaces the next one if we only want to take the positive and negative parts
    #Xp=X[int((N_tau-1)/2):N_tau,:]
    #Yp=Y_freq[int((N_tau-1)/2):N_tau,:]
    #Zp=(Z_RK_FT)[int((N_tau-1)/2):N_tau,:]

    #Xn=X[0:int((N_tau-1)/2),:]
    #Yn=Y_freq[0:int((N_tau-1)/2),:]
    #Zn=(Z_RK_FT)[0:int((N_tau-1)/2),:]

    # This part is when we set the negative part equal to zero
    Xp=X.copy()
    Yp=Y.copy()
    Zp=Z_RK_FT.copy()
    
    # Define the rotating frequency
    v_omega_rotating = v_omega_2_broad
    
    # Reshift the axes
    Y_shift = Yp.copy()-v_omega_rotating*hbar
    X_shift = Xp.copy()-v_omega_2_narrow*hbar-v_omega_1*hbar    
    Y_rescaled = Y_shift.copy() # prepare the array to be rescaled
    
    # XXXX XXXX
    # This was for the wrong sign of the phase factor e^i omega_2 tau, I had it with a negative sign 
    # for aa in range(len(grid_Ef)):
    #    Y_rescaled[:,aa] = Y_shift[:,aa]+X_shift[:,aa]
    
    # This is a trial
    for aa in range(len(grid_Ef)):
        Y_rescaled[:,aa] = -(Y_shift[:,aa] - X_shift[:,aa])
        
    
    
    # Final axes to be used are X_shift and Y_rescaled
    
    #######################################################################################
    # 3.2 Plot rescaled Fourier transform
    ####################################################################################### 
    
    #######################################################################################
    # 3.3 Return arrays
    #######################################################################################
    
    array_fourier_rescale  = [X_shift,Y_rescaled, Z_RK_FT, Z_RK_FT_rs_analytical_mod, 
                              spectrum_correction_analytical]
    plots_fourier_rescale  = [ax_fourier_transform_1,ax_fourier_transform_2,ax_fourier_transform_3]
    return_fourier_rescale = [array_fourier_rescale, plots_fourier_rescale]
        
    
    #######################################################################################
    # 4. INTERPOLATE
    #######################################################################################
    
    #######################################################################################
    # 4.1 Create interpolated grids
    #######################################################################################
    
    # Interpolation grids
    
    x_inter_lower = -0.79
    x_inter_upper =  0.79
    y_inter_lower = -0.79
    y_inter_upper =  0.79
    
    #x_inter_lower = np.min(X_shift)
    #x_inter_upper = np.max(X_shift)
    #y_inter_lower = x_inter_lower
    #y_inter_upper = y_inter_upper
    
    x_interpolation = np.linspace(x_inter_lower,x_inter_upper,N_interpolation)
    y_interpolation = np.linspace(y_inter_lower,y_inter_upper,N_interpolation)

    X_interpolate, Y_interpolate = np.meshgrid(x_interpolation, y_interpolation)

    # sigma_1_modulation
    sigma_1_modulation_around_zero = np.exp( -(Y_interpolate)**2/(2*hbar**2*v_sigma_1**2))
    
    
    # Interpolate the y direction. Make the function. 
    
    Xp2 = X_shift.copy()
    Yp2 = Y_rescaled.copy()    
    
    Zp2 = Z_RK_FT.copy()
    Zp2[np.isnan(Zp2)] = 0
        
    idx_min = find_nearest(Xp2[0,:], -0.8)
    idx_max = find_nearest(Xp2[0,:], +0.8)
    llist = np.arange(idx_min,idx_max+1,1) 
    
    ####################################################################
    
    # Interpolate the Z axis
    
    #Z_RK_FT_rs_analytical
    #Z_RK_FT_rs_intuitive
    #Z_RK_FT_rs_analytical_mod
    #Z_RK_FT_rs_intuitive_mod
    
    length = idx_max-idx_min+1
    N_new = 101
    
    
    inter_temp= np.zeros((length,N_new),'complex')
    inter_temp_1= np.zeros((length,N_new),'complex')
    inter_temp_2= np.zeros((length,N_new),'complex')
    inter_temp_3= np.zeros((length,N_new),'complex')
    inter_temp_4= np.zeros((length,N_new),'complex')
    Z_inter_RK_FT = np.zeros((N_new,N_new),'complex')
    Z_inter_RK_FT_rs_analytical = np.zeros((N_new,N_new),'complex')
    Z_inter_RK_FT_rs_analytical_mod = np.zeros((N_new,N_new),'complex')
    Z_inter_RK_FT_rs_intuitive = np.zeros((N_new,N_new),'complex')
    Z_inter_RK_FT_rs_intuitive_mod = np.zeros((N_new,N_new),'complex')
    Z_inter_RK_FT = np.zeros((N_new,N_new),'complex')
    
    T1 = Z_RK_FT_rs_analytical.copy()
    T2 = Z_RK_FT_rs_analytical_mod.copy()
    T3 = Z_RK_FT_rs_intuitive.copy()
    T4 = Z_RK_FT_rs_intuitive_mod.copy()
    
    
    new_x = Xp2[0,idx_min:idx_max+1]
    
    for aa in llist:
        x_temp = Yp2[:,aa]
        
        y_temp= Zp2[:,aa] # as a reminder Zp2 is Z_RK_FT or the FT without rescaling
        y_temp_1 = T1[:,aa]
        y_temp_2 = T2[:,aa]
        y_temp_3 = T3[:,aa]
        y_temp_4 = T4[:,aa]
        
        f = interp1d(x_temp,y_temp)
        f1 = interp1d(x_temp,y_temp_1)
        f2 = interp1d(x_temp,y_temp_2)
        f3 = interp1d(x_temp,y_temp_3)
        f4 = interp1d(x_temp,y_temp_4)
        
        inter_temp[aa-idx_min,:] = f(x_interpolation)
        inter_temp_1[aa-idx_min,:] = f1(x_interpolation)
        inter_temp_2[aa-idx_min,:] = f2(x_interpolation)
        inter_temp_3[aa-idx_min,:] = f3(x_interpolation)
        inter_temp_4[aa-idx_min,:] = f4(x_interpolation)
        
   
    for bb in range(N_new-1):
        fb = interp1d(new_x,inter_temp[:,bb])
        f1b = interp1d(new_x,inter_temp_1[:,bb])
        f2b = interp1d(new_x,inter_temp_2[:,bb])
        f3b = interp1d(new_x,inter_temp_3[:,bb])
        f4b = interp1d(new_x,inter_temp_4[:,bb])
        Z_inter_RK_FT[:,bb] = fb(x_interpolation)
        Z_inter_RK_FT_rs_analytical[:,bb] = f1b(x_interpolation)
        Z_inter_RK_FT_rs_analytical_mod[:,bb] = f2b(x_interpolation)
        Z_inter_RK_FT_rs_intuitive[:,bb] = f3b(x_interpolation)
        Z_inter_RK_FT_rs_intuitive_mod[:,bb] = f4b(x_interpolation)
        
        
    ####################################################################
        
    
    ####################################################################
    
    
    restore_profile_around_zero =  np.exp(-( (Y_interpolate)**2/(2*v_sigma_1**2)) )
    
    
    
    ####################################################################
    #### Compare with model
    ####################################################################
    
    rho_predicted = predicted_rho(params_system,params_protocol,params_simulation)
    rho_predicted = np.conjugate(rho_predicted)
    
 
    
    ###############
    
    axes_interpolated_ = [ax_interpolated_1,ax_interpolated_2,ax_interpolated_3,
                          ax_interpolated_4,ax_interpolated_5,ax_interpolated_6,
                          ax_interpolated_7,ax_interpolated_8,ax_interpolated_9]
    
    for aaa in range(9):
        axes_interpolated_[aaa].set_xlabel('energy (eV)',fontsize=22)
        axes_interpolated_[aaa].set_ylabel('energy (eV)',fontsize=22)
    
    axes_interpolated_[0].set_title('Unnormalized',fontsize=22)
    axes_interpolated_[1].set_title('Normalized',fontsize=22)
    axes_interpolated_[2].set_title('Predicted',fontsize=22)
    
    fig_interpolated.tight_layout()
    
    ### FIDELITY
    
    fid = fidelity(Z_inter_RK_FT,rho_predicted)
    fid_sigma_2_corrected = fidelity(Z_inter_RK_FT_rs_analytical_mod,rho_predicted)
    square = np.dot(Z_inter_RK_FT,Z_inter_RK_FT)
    square_corrected = np.dot(Z_inter_RK_FT_rs_analytical_mod,Z_inter_RK_FT_rs_analytical_mod)
    purity = np.trace(square)/np.trace(Z_inter_RK_FT)**2
    purity_sigma_2_corrected = np.trace(square_corrected)/np.trace(Z_inter_RK_FT_rs_analytical_mod)**2
    
    
    #######################################################################################
    # 4.2 Return arrays
    #######################################################################################
    
    array_interpolate = [X_interpolate, Y_interpolate, Z_inter_RK_FT,Z_inter_RK_FT_rs_analytical_mod, rho_predicted]
    plots_interpolate  = [ax_interpolated_1,ax_interpolated_2,ax_interpolated_3,
                         ax_interpolated_4,ax_interpolated_5,ax_interpolated_6,
                         ax_interpolated_7,ax_interpolated_8,ax_interpolated_9]
    return_interpolate = [array_interpolate, plots_interpolate]
    
    return_fidelity = [fid,fid_sigma_2_corrected,purity,purity_sigma_2_corrected]    
    

    #######################################################################################
    # Return
    #######################################################################################
    
    return [grid_Ef, freq, X_Ef, Y_tau, Y_freq, Z_RK_total, Z_RK_FT, Z_RK_FT_rs_analytical, Z_RK_FT_rs_intuitive,
           Z_RK_FT_rs_analytical_mod, Z_RK_FT_rs_intuitive_mod, spectrum_correction_analytical, 
           spectrum_correction_intuitive, mod,
           X_shift,Y_rescaled, X_interpolate, Y_interpolate, Z_inter_RK_FT,Z_inter_RK_FT_rs_analytical,
           Z_inter_RK_FT_rs_analytical_mod, Z_inter_RK_FT_rs_intuitive, Z_inter_RK_FT_rs_intuitive_mod,
           rho_predicted,fid,fid_sigma_2_corrected,purity,purity_sigma_2_corrected]
    
    
    
    #return  [return_time_traces, return_fourier_transform, return_fourier_rescale,return_interpolate,return_fidelity]

    
    
```
